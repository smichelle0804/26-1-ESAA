##### 시각화를 위한 차트 / 그래프 유형

종류 :
- 통계 분석을 위한 시각화 : 데이터 자체가 가지는 특성에 더 집중 
- 비즈니스 분석을 위한 시각화 : 청중 유형과 수준, 그리고 설득이 필요한 다양한 환경에서 사용 

##### 정보의 종류에 따른 시각화 차트 유형 
- 히스토 그램 : 연속형 데이터의 분포가 정규분포인지 왜곡되어 있는지 시각화 
- 바 플롯 : 특정 칼럼의 이산 값을 기준으로 다른 칼럼의 연속형 값을 막대 그래프로 시각화한다
- 박스 플롯 : 연속형 데이터의 분위를 시각화
- 바이올릿 플롯 : 연속형 데이터의 분포가 정규분포인지 왜곡되어 있는지 시각화 
- 스캐터 플롯(산점도) : 2개의 연속형 칼럼들에 대한 분포 및 관계를 점으로 시각화
- 상관 히트맵 : 다수의 연속형 칼럼들의 상관 관계를 온도를 나타내는 시각적인 컬러 기반으로 표현 <br>
바 플롯 제외한 모든 플롯들이 기본적으로 연속형 칼럼 값에 대한 시각화를 지원한다. <br>

시본 :
- X축, Y축으로 구성된 이차원 축에서 데이터를 시각화해주므로 2개의 변수(칼럼)에 대한 정보를 기본적으로 표출한다
- 시본의 대부분 시각화 함수들은 hue라는 인자를 통해서/플롯의 유형에 따라 2개의 변수가 아닌 3개의 변수도 함께 정보로 시각화 할 수 있다. 

##### 히스토그램 (Histogram)
- 사본뿐만 아니라 맷플로립, 판다스에서도 제공
- 어떠한 분포도를 가지고 있는지 확인하는 것
- 연속형 값을 범위 또는 구간으로 그룹화해 개별 구간에 해당되는 데이터의 건수를 시각화 해준다
- histplot() : Axes 레벨 함수
- displot() : Figure 레벨 함수 <br>
        - API 사용을 최소화 <br>
        - figure의 크기를 조절하기 위한 인자 : height(세로/높이의 크기), aspect(가로/세로 비율)<br>
        - width와 같은 별도로 가로(너비) 크기를 설정하는 인자는 제공하지 않는다<br>
        - 세부적인 변경이 다소 어렵다 <br>
        - 하지만 여러 시각화 함수들을 한 번에 시각화할 수 있다는 장점이 있다

```python
import pandas as pd

titanic_df = pd.read_csv("/Users/seomichelle/26-1 ESAA:Python/datasets/titanic_train.csv")
titanic_df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시본에서도 plt.figure()의 figsize 인자를 입력하여 전체 figure의 크기 조절
plt.figure(figsize=(8,3))

# Dataframe의 칼럼명을 자동으로 인식해서 xlabel 값을 할당, ylabel 값은 Count로 설정
sns.histplot(titanic_df['Age'], bins=20)
plt.show()
```

```python
sns.histplot(x='Age', data=titanic_df, bins=30, kde=True)
plt.show()

# bins=30 : 30개의 구간별 막대 그래프 도수가 그려짐
# kde = True에 따라 연속 확률분포 곡선까지 그려짐 
```

```python
import seaborn as sns
# seaborn의 figure레벨 그래프는 plt.figure()로 Figure 크기를 조절할 수 없다
plt.figure(figsize=(8,4))
sns.displot(x='Age', data=titanic_df, kde=True)
plt.show()
```

```python
sns.displot(titanic_df['Age'], kde=True, height=4, aspect=2)
plt.show
```

##### 카운트 플롯

- 카테고리성 칼럼 값별 건수를 시각화하는 데 사용한다
- 시본의 countoplot() 사용

```python
sns.countplot(x='Pclass', data=titanic_df)
plt.show
```

##### 바 플롯(barplot)
- 막대 그래프 형태의 플롯
- X축 값이 이산형 값으로 값이ㅢ 종류가 너무 많지 않을 때 유용하게 사용될 수 있다
- barplot() 
- 수직 또는 수평 막대 그래프를 시각화 선택을 orient인자로 설정할 수 있다
- orient가 v인 경우 수직(vertical), h일 경우는 수평(hoizontal)
- orient 값을 별도로 설정하지 않을 경우 barplot()함수는 입력된 X,Y축 값의 데이터 유형을 판단하여 자동으로 수직 또는 수평 막대 그래프를 그려준다
- X,Y축 모두 이산값을 입력한다면 바 플롯으로 표현되는 정보를 시각적으로 잘못 이해될 수 있다

```python
# 자동으로 xlabel, ylabel을 barplot()의 x 인자값인 Pclass, y 인자값이 Age로 설정 
sns.barplot(x='Pclass', y='Age', data=titanic_df)
plt.show
```

```python
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.show()
```

```python
# x 인자로 문자열인 Name을, y 인자로 문자열인 Sex를 입력하므로 barplot은 오류를 발생 
sns.barplot(x='Name', y='Sex', data=titanic_df)
```

```python
# estimator = sum을 적용하여 평균이 아니라 총합으로 표현
sns.barplot(x='Pclass', y='Survived', data=titanic_df, ci=None, estimator=sum)
plt.show()
```

##### barplot() 함수의 hue 인자를 사용하여 시각화 정보를 추가적으로 세분화하기
- hue 인자를 통해 세부적인 추가 정보를 더해서 시각화 할 수 있다

```python
# 아래는 Pclass가 X축값이며 hue파라미터를 Sex를 설정
# 개별 Pclass 값별로 Sex에 따른 Age 평균 값을 구함 
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic_df)
plt.show()
```

```python
# 개별 Pclass 값별로 Sex에 따른 Survived 평균값을 구함 
# Pclass가 X축이며 Survived가 Y축 값, hue 파라미터로 sex 설정
sns.barplot(x='Pclass',y='Survived', hue='Sex', data=titanic_df)
plt.show()
```

##### 바이올린 플롯
- 히스토그램의 연속 확률 분포 곡선과 박스 플롯을 함께 시각화 가능
- 연속형 값에 적용해야 의미 있는 정보를 시각화 가능
- violinplot()
- 이산값별로 여러 개의 히스토그램 연속 확률 불포 곡선들을 그래 줄 수 있다

```python
# x축 값인 Pclass 값별로 Y축 값인 Age의 바이올린 플롯을 그림
# pclass 값별 Age 데이터 불포를 비교하여 볼 수 있다
sns.violinplot(x='Pclass', y='Age', data=titanic_df)
plt.show()
```

```python
# x축 값이 개별 pclass 내에서 sex 값별로 y축 값인 age의 바이올린 플롯을 그림
sns.violinplot(x='Pclass', y='Age', hue='Sex', data=titanic_df)
plt.show()
```

##### subplots를 이용하여 시본의 다양한 그래프를 시각화
- Axes 객체 변수를 입력하게 되면 해당 위치의 subplot에 카운트 플롯을 시각화 해준다

```python
cat_columns = ['Survived', 'Pclass', 'Sex']

# nrows는 1이고 ncols는 칼럼의 개수만큼 subplots을 설정
fig, axs = plt.subplots(nrows=1, ncols=len(cat_columns), figsize=(14,4))

for index, column in enumerate(cat_columns):
    print('index:', index)
    # seaborn의 Axes 레벨 function들은 ax 인자로 subplots의 어느 Axes에 위치할지 설정
    sns.countplot(x=column, data=titanic_df, ax=axs[index])
```

```python
cat_columns = ['Pclass', 'Sex', 'Embarked']

# nrows는 1이고 ncols는 칼럼의 개수만큼인 subplots을 설정.
fig, axs = plt.subplots(nrows=1, ncols=len(cat_columns), figsize=(14, 4))

for index, column in enumerate(cat_columns):
    # seaborn의 Axes 레벨 function들은 ax 인자로 subplots의 어느 Axes에 위치할지 설정.
    sns.barplot(x=column, y='Survived', data=titanic_df, ax=axs[index])
```

```python
cont_columns = ['Age', 'Fare']

# 리스트로 할당된 칼럼들의 개수만큼 루프 수행.
for column in cont_columns:

    # 왼쪽에는 바이올린 플롯, 오른쪽에는 히스토그램을 시각화. nrows는 1, ncols=2인 서브플롯 생성.
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

    # 왼쪽 Axes 객체에는 Survived값 0, 1별 개별 칼럼의 바이올린 플롯 시각화.
    sns.violinplot(x='Survived', y=column, data=titanic_df, ax=axs[0])

    # 오른쪽 Axes 객체에는 Survived 값에 따른 개별 칼럼의 히스토그램 시각화
    sns.histplot(x=column, data=titanic_df, kde=True, hue='Survived', ax=axs[1])
```

##### 상관 히트맵(Correlation Heatmap)

- 다수의 속성들 간의 상관계수를 히트맵 형태로 나타낼 수 있다
- 수치값을 온도를 연상시키는 다양한 색상으로 표현
- 상관계수는 두 속성들 간의 선형적인 연관 관계를 수치화한 값이다 (두 값이 상관없으면 0, 반대면 -1, 같은 방향이면 1)
- heatmap() : 상관계수를 가지는 DataFrame을 입력받아야 시각화가 가능하다
- DataFrame의 corr() 메서드를 호출하면 간단하게 상관계수를 가지는 DataFrame을 생성할 수 있다
- cmap인자를 통해 히트맵의 색상을 변경할 수 있게 해준다
- annot인자는 T/F를 설정할 수 있으며 True일 경우 숫자로 된 상관계숫값을 표시해준다 <br>
    - annot 인자로 숫자 상관계숫값을 표시할 경우 정밀도가 너무 높으면 개별 숫자값이 길이가 커져서 서로 겹쳐 보이게 된다 --> fmt 인자로 숫자값의 포맷 변환 가능 
    - fmt='.1f' 설정시 소수점 한 자리까지 상관계숫값을 표시한다
    - cbar인자 : T/F 설정 가능, True일 경우 숫자값에 따른 색깔 기준 막대로 표시한다(기본)

```python
corr_df = titanic_df.corr(numeric_only=True)
corr_df
```

```python
# 상관계수를 DataFrame으로 구하고 이를 heatmap()의 인자로 입력
corr = titanic_df.corr(numeric_only=True)
sns.heatmap(corr)
plt.show()
```

```python
sns.heatmap(corr, annot=True,fmt='.1f', cbar=True)
plt.show
```
