### 02. 맷플롯립 (Matplotlib)

##### 맷플로립의 pyplot 모듈의 이해
- 파이썬 시각화를 위한 기반 모듈인 pyplot을 제공하며 이를 통해 시각화를 할 수 있다.
- plot() : 함수를 통해 선 그래프를 그릴 수 있다
- title() : 입력값으로 타이틀 문자열을 받아서 차트 타이틀 설정 가능
- show() : 축을 포함한 선 그래프를 화면에 나타낼 수 있다



```python
# pyplot 모듈을 (관용적으로) plt라는 이름으로 import
import matplotlib.pyplot as plt

# x축 입력값 [1,2,3], y축 입력값 [2,4,6]으로 선 그래프 생성
plt.plot([1,2,3], [2,4,6])

#선 그래프 타이틀을 Hello plot으로 설정
plt.title("Hello plot")

#선 그래프 출력
plt.show()
```

##### pyplot의 두 가지 중요 요소 - Figure와 Axes의 이해

Figure
- 그림을 그리기 위한 캔버스 역할
- 크기를 조절, 플롯을 최종적으로 화면에 나타내는 역할 수행
- 하지만 실제적으로 그림을 그리는 역할은 하지 않는다

Axes
- 실제 그림을 그리는 메서드
- 대부분의 시각화를 위해 호출됨
- X축 , Y축, 타이틀, 범례 등의 속성을 설정하는 데도 이용된다

##### Figure와 Axis의 활용

figsize = (가로크기, 세로크기)

plt.figure()는 주로 figure의 크기를 조절하는 데 사용된다
- 함수의 인자로 facecolor= 을 통해 그림판의 색을 변경할 수 있다

```python
plt.figure(figsize=(10,4))

plt.plot([1,2,3], [2,4,6])
plt.title("Hello plot")
plt.show()
```

```python
figure = plt.figure(figsize=(10,4))
print(type(figure))
```

```python
plt.figure(figsize=(4,4), facecolor='yellow')

plt.plot([1,2,3], [2,4,6])
plt.title("Hello plot")
plt.show()
```

plt의 axes() 함수는 현재 사용하는 Axes 객체를 반환한다
- plt.axes()로 반환된 객체 AxesSubplot는 Axes의 실체화된 하위 클래스다

plt.subplots() : pyplot에서 Figure와 Axes 객체를 함께 가져올 수 있다
- subplots() 함수는 인자로 여러개의 Axes를 설정할 수 있다

```python
ax = plt.axes()
print(type(ax))
```

```python
### pyplot에서 설정된 Figure와 Axes 객체를 함께 가져오기
fig, ax = plt.subplots()
print('fig type:', type(fig), '\nax type:', type(ax))
```

```python
fig, ax = plt.subplots(figsize=(4,4))

#Axes 객체의 plot()을 이용하여 선 그래프를 그림
ax.plot([1,2,3], [2,4,6])

# Axes 객체의 set_title()을 이용하여 선 그래프 타이틀을 Hello plot으로 설정
ax.set_title('Hello plot')
plt.show()
```

##### 여러 개의 plot을 가지는 subplot들을 생성하기

plt.subplots() : 여러개의 subplot을 생성하는 데 활용한다
- 여러 개의 그래프를 그릴 수 있게 하는 1개의 figure와 여러 개의 Axes를 생성하고 반환한다.
- 인자 : nrows , ncols, figsize

nrows : 전체 subplot들의 배치를 2차운 행렬 형태로 표현할 때 행의 개수가 된다 <br>
ncols : 열의 개수 <br>
figsize : 모든 subplot들을 포함한 전체 Figure의 크기를 설정

```python
# 아래의 두 개의 subplot, 즉 두 개의 Axes를 생성하되 열 방향으로 두 개를 배치함
# 또한 전체 크기가 가로6, 세로 3인 figure 생성
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
```

##### pyplot의 plot() 함수를 이용해 선 그래프 그리기

pyplot의 plot() : 선 그래프를 그릴 때 활용 <br>
- 파이썬 리스트, ndarray, 판다스의 DataFrame/Series 모두 적용 가능 
- 단 X좌표 값과 Y 좌표 값의 개수는 모두 같은 크기를 가져야 한다

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x_value = [1, 2, 3, 4]
y_value = [2, 4, 6, 8]

# x_value = np.array([1, 2, 3, 4])  # 넘파이 array로 테스트하려면 주석을 해제하고 실행
# y_value = np.array([2, 4, 6, 8])  # 넘파이 array로 테스트하려면 주석을 해제하고 실행

# 입력값으로 파이썬 리스트, numpy array 가능. 
# x 좌표값과 y 좌표값은 모두 같은 크기를 가져야 함.

plt.plot(x_value, y_value)
plt.show()
```

```python
# color 인자를  green으로 설정하여 선의 색깔을 녹생으로 변경
plt.plot(x_value, y_value, color='green')
```

```python
#API 기반으로 시각화를 구현할 때는 함수의 인자들에 대해 알고 있어야 하는 부작용(?)이 있다
plt.plot(
    x_value, y_value, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12
)
```

축의 눈금값(틱값) 회전 
- 이유 : 축의 값이 문자열이고 많은 눈금값이 있을 경우 눈금 간에 값이 중복되면서 눈금값이 제대로 보이지 않기 때문

xticks() : x축 눈금값 회전 <br>
yticks() : y축 눈금값 회전 

```python
# x 좌표 값을 0~99까지 100개로 설정
x_value = np.arange(0,100)

# y 좌표 값은 개별 x좌표값을 2배로 해서 100개 설정
y_value = 2*x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

# x축 눈금값을 90도로 회전
plt.xticks(rotation=90)
plt.show()
```

범례(legend) 설정
- 여러 개의 그래프/차트를 한 Axes 내에서 표시할 때 이들을 구분하기 편리하게 위해


```python
import numpy as np
import matplotlib.pyplot as plt

# X좌표값은 0~99까지 100개
# 첫 번째 Y좌표값은 X좌표값의 두 배, 두 번째 좌표값은 X좌표값의 네 배
x_value_01 = np.arange(0, 100)

y_value_01 = 2 * x_value_01
y_value_02 = 4 * x_value_01

# 녹색 선 그래프
plt.plot(x_value_01, y_value_01, color='green', label='green line')

# 빨간색 선 그래프
plt.plot(x_value_01, y_value_02, color='red', label='red line')

plt.xlabel('x axis')
plt.ylabel('y axis')

# 범례 표시
plt.legend()

plt.title('Hello plot')

plt.show()
```

```python
x_value_01 = np.arange(1, 10)
y_value_01 = 2 * x_value_01

# 마커가 있는 빨간색 점선 그래프
plt.plot(x_value_01, y_value_01,
         color='red',
         marker='o',
         linestyle='dashed',
         label='red line')

# 초록색 막대 그래프
plt.bar(x_value_01, y_value_01,
        color='green',
        label='bar plot')

plt.xlabel('x axis')
plt.ylabel('y axis')

# 범례 표시
plt.legend()

plt.title('Hello plot')

plt.show()
```

위 예제와 동일한 시각화 결과를 출력하되, 직접 Axes 객체 메서드를 사용해 시각화를 수행 

```python
figure = plt.figure()
ax = plt.axes()

ax.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax.bar(x_value_01, y_value_01, color='green', label='bar plot')

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax.legend()
ax.set_title('Hello plot')

plt.show()
```

##### 여러 개의 subplot들을 이용해 개별 그래프들을 subplot별로 시각화하기

여러 개의 그래프들을 한꺼번에 시각화 가능 --> 데이터 분석 측면에서 매우 유용 <br>
유의사항 : 개별 Axes 객체를 직접 이요해서 시각화를 구형해야한다. 

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
```

```python
print('ax type:', type(ax))
print('ax[0] type', type(ax[0]))
print('ax[1] type:', type(ax[1]))
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)

y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_02, y_value_02, color='green', label='bar plot')

ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

ax[0].legend()
ax[1].legend()

ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
```

```python
# figsize는 (6,6)인 Figure와 2x2개의 Axes를 가지는 subplot을 반환
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

# 2x2 Axes 객체들을 각각 이용하여 선 그래프와 막대그래프를 그림
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='green bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')

# 개별 Axes 객체 각각에 x축명을 설정
ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

# 개별 Axes 객체 각각에 범례 설정
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```
