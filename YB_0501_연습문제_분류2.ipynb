```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('/Users/seomichelle/26-1 ESAA:Python/datasets/와인/train.csv')
test = pd.read_csv('/Users/seomichelle/26-1 ESAA:Python/datasets/와인/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())

import pandas as pd
from sklearn.preprocessing import LabelEncoder

train_one = pd.get_dummies(data = train, columns = ['type'], prefix = ['type'])

test_one = pd.get_dummies(data = test, columns = ['type'], prefix = ['type'])

encoder = LabelEncoder()
train_one['quality'] = encoder.fit_transform(train_one['quality'])

train_one.head()

```

```python
# 모델 정의
model = XGBClassifier()
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수
X = train_one.drop('quality',axis= 1)
y = train_one['quality']

# fit 메소드를 이용해 모델 학습
model.fit(X,y)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측
pred = model.predict(test_one)
pred
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
submission = pd.read_csv('/Users/seomichelle/26-1 ESAA:Python/datasets/와인/sample_submission.csv')
submission['quality'] = pred + 3ㅇ
submission.head()
```

```python
# 정답파일 내보내기
submission.to_csv('xgb_pred.csv',index = False)
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('/Users/seomichelle/26-1 ESAA:Python/datasets/와인/train.csv')
test = pd.read_csv('/Users/seomichelle/26-1 ESAA:Python/datasets/와인/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train_one = pd.get_dummies(train)
test_one = pd.get_dummies(test)
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
LGBM = LGBMClassifier()
XGB = XGBClassifier()
RF = RandomForestClassifier()

# VotingClassifier 정의
VC = VotingClassifier(estimators=[('rf',RF),('xgb',XGB),('lgbm',LGBM)],voting = 'soft')
```
